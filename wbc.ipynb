```python
import pandas as pd, numpy as py, matplotlib.pyplot as plt
```

```python
df=pd.read_csv('wbc.csv')
```

```python
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.model_selection import train_test_split as tts
```

```python
dt=dtc()
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=tts(df.drop('diagnosis',axis=1),df.diagnosis,test_size=0.09,random_state=32)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#gini
print(dt)
dt.fit(X_train,y_train)
dt.score(X_test,y_test)
```

```python
dt=dtc(criterion='entropy')
```

```python
#entropy
print(dt)
dt.fit(X_train,y_train)
dt.score(X_test,y_test)
```

### Clearly for this type of data 'entropy' is more effective than 'gini'
